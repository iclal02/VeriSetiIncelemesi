```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("Crime_Data_from_2020_to_Present.csv",encoding='utf-8')
```

```python
df.head()
```

```python
# Part 1-2 sütununun veri türünü ve benzersiz değerleri kontrol et part-1 ağır suç part-2 daha az ağır
print(df['Part 1-2'].value_counts())
```

```python
# Bölge ve suç türü (Part 1-2) bazında gruplandırma ve suç sayısını hesaplama
area_crime_distribution = df.groupby(['AREA NAME', 'Part 1-2']).size().unstack(fill_value=0)
area_crime_distribution.columns=["Ağır suç","Hafif suç"]

# Gruplandırılan veriyi incele
print(area_crime_distribution.head())
```

```python
# Yığınlı bar grafiği oluştur
#stacked birbirinin üstüne gelmesini sağlar,viridis renk skalasıdır
area_crime_distribution.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='viridis')
plt.title('Bölgelere Göre Part 1 ve Part 2 Suçlarının Dağılımı')
plt.xlabel('Bölge (AREA)')
plt.ylabel('Suç Sayısı')
plt.legend(title='Suç Türü')
plt.grid(True)
plt.show() 
#burdan hangi bölgelerde suçun ne oranda oldugunu öğrenebiliriz mesela foothill de suç oranı daha az en yüksek suç oranı centralda
```

SUÇLARIN GERÇEKLEŞTİĞİ SAATLER

```python
# TIME OCC sütununu saat ve dakika olarak bölme
df['Hour'] = df['TIME OCC'] // 100  # Saat kısmını alır
df['Minute'] = df['TIME OCC'] % 100  # Dakika kısmını alır

# Her saatteki suç sayısını gruplama
crime_by_hour = df.groupby('Hour').size()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(crime_by_hour.index, crime_by_hour.values, color='blue')
plt.title('Saat Bazında Suç Dağılımı')
plt.xlabel('Saat (24 saat formatı)')
plt.ylabel('Suç Sayısı')
plt.grid(True)
plt.show()
```

SUÇLARIN İŞLENME ÇOKLUĞU

```python
#şimdi de suçlara bakalım
#size her grup için toplam sayıyı gösterir
Crime_Grouping=df.groupby('Crm Cd Desc').size().reset_index(name='Count')
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(x='Crm Cd Desc', y='Count', data=Crime_Grouping, palette='viridis')
plt.title('Suç Gruplaması Dağılımı', fontsize=16)
plt.xlabel('Suç Türleri', fontsize=14)
plt.ylabel('Toplam Sayı', fontsize=14)
plt.xticks(rotation=45)  # X eksenindeki etiketleri döndür
plt.show()
# çok fazla veri oldugundan karıstı bu yuzden biz max degerdekileri alalım
```

```python
#genelde en çok işlenen suç hangisiymiş araç çalınmasıymış
top_10_crimes = Crime_Grouping.nlargest(10, 'Count')
top_10_crimes
```

```python
#şimdi merkezde gerçekleşen en çok suç türünü bulalım

# Merkez alanını filtrele
central_area_df = df[df['AREA NAME'] == 'Central']

# Merkezdeki suç türlerine göre gruplama ve sayma
#reset_index olmayan bir sütun oluşturur
central_crime_grouping = central_area_df.groupby('Crm Cd Desc').size().reset_index(name='Count')
central_crime_grouping
```

```python
# En sık görülen suç türünü bul
most_common_crime = central_crime_grouping.loc[central_crime_grouping['Count'].idxmax()]

print("Merkezde en çok görülen suç türü:")
print(most_common_crime)
#araçtan hırsızlık en çok işlenen suçmuş
```

```python
#şimdi foothillde gerçekleşen en çok suç türünü bulalım

# foothill alanını filtrele
foothill_area_df = df[df['AREA NAME'] == 'Foothill']

# foothilldeki suç türlerine göre gruplama ve sayma
foothill_crime_grouping = foothill_area_df.groupby('Crm Cd Desc').size().reset_index(name='Count')
foothill_crime_grouping
# En sık görülen suç türünü bul
most_common_crime_foothill = foothill_crime_grouping.loc[foothill_crime_grouping['Count'].idxmax()]

print("Foothillde en çok görülen suç türü:")
print(most_common_crime_foothill) #araç hırsızlığı en çok işlenen suçmuş
```

```python
# "VEHICLE - STOLEN" suç türünü filtrele
burglary_data = df[df['Crm Cd Desc'] == 'VEHICLE - STOLEN']
burglary_count_by_area = burglary_data.groupby('AREA NAME').size().reset_index(name='Count')

plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='AREA NAME', data=burglary_count_by_area, palette='viridis')
plt.title('Bölgelerde Araç Çalınımı Suçlarının Dağılımı', fontsize=16)
plt.xlabel('Toplam Suç Sayısı', fontsize=14)
plt.ylabel('Bölgeler', fontsize=14)
plt.show()
```

```python
# En yüksek iki değeri al
top_two_burglary = burglary_count_by_area.nlargest(2, 'Count')
top_two_burglary
```
