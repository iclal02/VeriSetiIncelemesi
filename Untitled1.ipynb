```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("Crime_Data_from_2020_to_Present.csv",encoding='utf-8')
```

```python
df.head()
```

```python
df.columns
```

```python
# Date Rptd sütununu tarih formatına çevirdik ve yılı bulduk
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p')
df['Year'] = df['Date Rptd'].dt.year
```

```python
# 2023 yılı ve 20-60 yaş aralığındaki kurbanları filtreledik
filtered_df = df[(df['Year'] == 2023) & (df['Vict Age'].between(20, 60))]
```

```python
# Bölge ve silah kodu kombinasyonunun toplam kurban sayısını analiz ettik
heatmap_data = filtered_df.groupby(['AREA NAME', 'Weapon Used Cd', 'Vict Sex']).size().unstack(fill_value=0)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu')
plt.title('Kurbandan Silah Kullanımı ve Cinsiyete Göre Heatmap')
plt.ylabel('Bölge Adı')
plt.xlabel('Kullanılan Silah Kodu')
plt.show()
```

en yüksek kurban sayısına sahip olan 3 bölgeyi belirledik ve bu bölgelerde kullanılan silahların türlerine göre  dağılımını bulduk.


```python
# En yüksek kurban sayısına sahip 3 bölgeyi belirledik
top_areas = filtered_df['AREA NAME'].value_counts().head(3)
print(f"En yüksek kurban sayısına sahip 3 bölge:\n{top_areas}\n")
```

```python
# Bu bölgelerde kullanılan silahların türlerine göre dağılımlarını bulduk
top_area_names = top_areas.index
top_weapons = filtered_df[filtered_df['AREA NAME'].isin(top_area_names)]['Weapon Used Cd'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_weapons, labels=top_weapons.index)
plt.title('En Yüksek Kurban Sayısına Sahip Bölgelerde Kullanılan Silah Türleri')
plt.axis('equal') #dairesel görünümü sağlar
plt.show()
```

cinsiyete göre en sık karşılaşılan silahları ve kurbanların yaş ortalamasını hesaplayarak, 
cinsiyet ve silah kullanımı arasında bir ilişki olup olmadığını inceledik.
Özellikle erkek kurbanların kullandığı silahların yaş aralığına göre dağılımına baktık.

```python
#Cinsiyete göre en sık karşılaşılan silahları ve yaş ortalamasını hesapladık
gender_weapon_analysis = filtered_df.groupby(['Vict Sex', 'Weapon Used Cd']).size().unstack(fill_value=0)
average_age = filtered_df['Vict Age'].mean()

print(f"Cinsiyete Göre En Sık Karşılaşılan Silahlar:\n{gender_weapon_analysis}\n")
print(f"2023'teki kurbanların yaş ortalaması: {average_age:.2f}")
```

```python
# Erkek kurbanların kullandığı silahların yaş aralığına göre dağılımı
male_filtered = filtered_df[filtered_df['Vict Sex'] == 'M']
male_age_distribution = male_filtered['Weapon Used Cd'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=male_age_distribution.index, y=male_age_distribution.values)
plt.title('Erkek Kurbanların Kullandığı Silahların Dağılımı')
plt.xlabel('Kullanılan Silah Kodu')
plt.ylabel('Kurban Sayısı')
plt.xticks(rotation=45)
plt.show()
```
